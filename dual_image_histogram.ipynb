
# Dual‑Image Histogram Viewer

Load two images and visualize their **grayscale intensity histograms** together on a single chart.


````python

%matplotlib inline
from pathlib import Path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['axes.grid'] = True
````

````python

def load_grayscale(image):
    """Return grayscale PIL.Image (mode 'L')."""
    if isinstance(image, Image.Image):
        return image.convert('L')
    return Image.open(Path(image)).convert('L')
````

````python

def plot_two_histograms(img1, img2, labels=('Image 1', 'Image 2')):
    """Overlay grayscale histograms of two images."""
    im1 = load_grayscale(img1)
    im2 = load_grayscale(img2)
    data1 = np.array(im1).ravel()
    data2 = np.array(im2).ravel()
    
    bins = np.arange(257)
    plt.figure()
    plt.hist(data1, bins=bins, alpha=0.5, label=labels[0], density=True, histtype='stepfilled')
    plt.hist(data2, bins=bins, alpha=0.5, label=labels[1], density=True, histtype='step')
    plt.xlabel('Pixelintensität (0-255)')
    plt.ylabel('Normalisierte Frequenz')
    plt.title('Graustufenhistogramme')
    plt.legend()
    plt.tight_layout()
    plt.show()
````

````python
def plot_two_histogram(img1, label='Image'):
    """Overlay grayscale histograms of two images."""
    im1 = load_grayscale(img1)
    data1 = np.array(im1).ravel()
    
    bins = np.arange(257)
    plt.figure()
    plt.hist(data1, bins=bins, alpha=0.5, label=label, density=True, histtype='stepfilled')
    plt.xlabel('Pixelintensität (0-255)')
    plt.ylabel('Normalisierte Frequenz')
    plt.title('Graustufenhistogramm')
    plt.legend()
    plt.tight_layout()
    plt.show()
````

````python
def plot_three_histograms(img1, img2, img3, labels=('Image 1', 'Image 2', 'Image 3')):
    """Overlay grayscale histograms of two images."""
    im1 = load_grayscale(img1)
    im2 = load_grayscale(img2)
    im3 = load_grayscale(img3)
    data1 = np.array(im1).ravel()
    data2 = np.array(im2).ravel()
    data3 = np.array(im3).ravel()
    
    bins = np.arange(257)
    plt.figure()
    plt.hist(data1, bins=bins, alpha=0.5, label=labels[0], density=True, histtype='stepfilled')
    plt.hist(data2, bins=bins, alpha=0.5, label=labels[1], density=True, histtype='step')
    plt.hist(data3, bins=bins, alpha=0.5, label=labels[2], density=True, histtype='step')
    plt.xlabel('Pixelintensität (0-255)')
    plt.ylabel('Normalisierte Frequenz')
    plt.title('Graustufenhistogramme')
    plt.legend()
    plt.tight_layout()
    plt.show()
````
