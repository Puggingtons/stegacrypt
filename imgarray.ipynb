```python
# Cell 1  imports
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg  # or use PIL.Image if you prefer

# Cell 2  collect or list your eight images
# Option A: specify them manually in the order you want
image_paths = [
    "dhbw_stegacrypt_1.png",
    "dhbw_stegacrypt_2.png",
    "dhbw_stegacrypt_3.png",
    "dhbw_stegacrypt_4.png",
    "dhbw_stegacrypt_5.png",
    "dhbw_stegacrypt_6.png",
    "dhbw_stegacrypt_7.png",
    "dhbw_stegacrypt_8.png",
]

# Option B: or gather the first eight files that match a pattern
# image_paths = sorted(glob.glob("outputs/quality_variants/*.png"))[:8]

assert len(image_paths) == 8, "Need exactly eight input images"

# Cell 3  build the figure
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 8))
for idx, (ax, img_path) in enumerate(zip(axes.flatten(), image_paths), start=1):
    ax.imshow(mpimg.imread("examples/output/" + img_path))
    ax.set_title(f"Anzahl der Bit-Ebenen: {idx}", fontsize=10)
    ax.axis("off")          # hide tick marks and axes

plt.tight_layout()
plt.show()

# Optionally save the combined overview
# fig.savefig("8-qualities_overview.png", dpi=300, bbox_inches="tight")
```
